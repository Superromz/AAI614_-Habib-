This notebook demonstrates simple data visualization in Python using `matplotlib` and `plotly`.  
We’ll generate a random dataset, explore it, and visualize it both statically and interactively.

We’ll need `numpy` and `pandas` for data handling, `matplotlib` for static plotting, and `plotly` for interactive visualization.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

Let’s simulate a simple random walk using NumPy. This is a common technique for generating test data in data science and finance.
python
Copy
Edit



```python
# Set random seed for reproducibility
np.random.seed(42)

# Create 100 steps
steps = 100

# Generate random walk data
random_steps = np.random.normal(loc=0, scale=1, size=steps)
random_walk = np.cumsum(random_steps)

# Build DataFrame
df = pd.DataFrame({
    "Step": np.arange(1, steps + 1),
    "Value": random_walk
})

df.head()
```

Before plotting, let’s explore our generated data. We'll look at the first few rows, some basic stats, and a simple histogram.
python
Copy
Edit


```python
# Show the first five rows
print(df.head())

# Basic statistics
print(df.describe())

# Histogram
df['Value'].hist(bins=15, figsize=(8,4))
plt.title('Distribution of Random Walk Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

Let’s create a line plot to visualize how the random walk changes over time.
python
Copy
Edit


```python
plt.figure(figsize=(12,6))
plt.plot(df['Step'], df['Value'], marker='o', linestyle='-', color='teal')
plt.title('Random Walk Over Time')
plt.xlabel('Step')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

## 5. Interactive Data Visualization with plotly

For an interactive experience, let’s use Plotly to make a scatter plot. Try zooming, hovering, and panning!


```python
fig = px.scatter(df, x='Step', y='Value', title='Random Walk (Interactive)')
fig.show()
```
